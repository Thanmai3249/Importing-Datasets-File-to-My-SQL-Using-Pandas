```python
#Create a table and Import the CSV data into the MySQL table

import mysql.connector as connection
from mysql.connector import Error

mydb = connection.connect(host='localhost', database='Superstore', user='root', password='thanmai123')
if mydb.is_connected():
    cursor = mydb.cursor()
    cursor.execute("select database();")
    record = cursor.fetchone()
    print("You're connected to database: ", record)
    cursor.execute('DROP TABLE IF EXISTS orders;')
    print('Creating table....')
# in the below line please pass the create table statement which you want #to create
cursor.execute('''create table orders(
                    RowID                          varchar(500),
                    OrderPriority                  varchar(500),
                    Discount                       int,
                    UnitPrice                     int,
                    ShippingCost                  int,
                    CustomerID                      int,
                    CustomerName                   varchar(500),
                    ShipMode                       varchar(500),
                    CustomerSegment                varchar(500),
                    ProductCategory                varchar(500),
                    ProductSubCategory            varchar(500),
                    ProductContainer               varchar(500),
                    ProductName                    varchar(500),
                    ProductBaseMargin            float(10,10),
                    Region                          varchar(500),
                    StateorProvince               varchar(500),
                    City                            varchar(500),
                    PostalCode                      int,
                    OrderDate              varchar(500),
                    ShipDate               varchar(500),
                    Profit                         int,
                    Quantityorderednew             int,
                    Sales                          int,
                    OrderID                         int
                    );
                    update orders
                    set OrderDate=date_format(OrderDate,"%d/%m/%y");
                    set ShipDate=date_format(ShipDate,"%d/%m/%y");
                    ''')
print("Table is created....")
```
